### Settings

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Billing dictionary
dict_billing = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'value': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
dict_billing
```

```python
df_billing = pd.DataFrame.from_dict(dict_billing)
```

```python
df_billing
```

```python
df_billing.describe()
```

```python
df_billing.info()
```

```python
# Convert DataFrame object to period
df_billing['data_ref'] = pd.to_datetime(df_billing['data_ref']).dt.to_period('M')
```

```python
df_billing.info()
```

### Bring the average sales

```python
# Bring the average sales
df_billing.value.mean()
```

```python
# Boxplot graph
df_billing.plot.box()
```

### Vertical bar chart showing the reference month and value

```python
# Vertical bar chart showing the reference month and value - without sort
df_billing.plot.bar(
    x='data_ref',
    y='value',
    color='#A020F0',
    )
```

```python
# Vertical bar chart showing the reference month and value - with sort
df_billing.sort_values('data_ref').plot.bar(
    x='data_ref',
    y='value',
    color='#FFA500',
    )
```

### Vertical line chart showing the reference month and value

```python
# Vertical line chart showing the reference month and value - without sort
df_billing.plot.line(
    x='data_ref',
    y='value',
    color='#A020F0',
    )
```

```python
# Vertical line chart showing the reference month and value - with sort
df_billing.sort_values('data_ref').plot.line(
    x='data_ref',
    y='value',
    color='#FFA500',
    )
```
